```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "/Users/emilynotaro/Downloads/Starter_Code 2/Pymaceuticals/data/Mouse_metadata.csv"
study_results_path = "/Users/emilynotaro/Downloads/Starter_Code 2/Pymaceuticals/data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_study_data = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Display the data table for preview
mouse_study_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_study_data = mouse_study_data.set_index("Mouse ID")
clean_mice = mouse_study_data.drop(index=("g989"))
clean_mice
```

```python
#Summary Statistics
```

```python
drug_names = clean_mice["Drug Regimen"].unique()
drug_names
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

per_drug_mean = clean_mice.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
per_drug_median = clean_mice.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
per_drug_var = clean_mice.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
per_drug_std = clean_mice.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
per_drug_sem = clean_mice.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

stats_by_drug = pd.merge(per_drug_mean, per_drug_median, on="Drug Regimen")
stats_by_drug = stats_by_drug.rename(columns={"Tumor Volume (mm3)_x": "Mean of Tumor Volume",
                                              "Tumor Volume (mm3)_y": "Median of Tumor Volume"})
stats_by_drug = pd.merge(stats_by_drug, per_drug_var, on="Drug Regimen")
stats_by_drug = stats_by_drug.rename(columns={"Tumor Volume (mm3)": "Variance of Tumor Volume"})
stats_by_drug = pd.merge(stats_by_drug, per_drug_std, on="Drug Regimen")
stats_by_drug = stats_by_drug.rename(columns={"Tumor Volume (mm3)": "Standard Deviation of Tumor Volume"})
stats_by_drug = pd.merge(stats_by_drug, per_drug_sem, on="Drug Regimen")
stats_by_drug = stats_by_drug.rename(columns={"Tumor Volume (mm3)": "Standard Error of the Mean (SEM) of Tumor Volume"})

stats_by_drug
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
stats_agg = clean_mice.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})
stats_agg
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

capomulin_only = clean_mice.loc[clean_mice["Drug Regimen"] == "Capomulin", :]
ramicane_only = clean_mice.loc[clean_mice["Drug Regimen"] == "Ramicane", :]
infubinol_only = clean_mice.loc[clean_mice["Drug Regimen"] == "Infubinol", :]
ceftamin_only = clean_mice.loc[clean_mice["Drug Regimen"] == "Ceftamin", :]

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint_capomulin = capomulin_only.groupby(["Mouse ID"]).max()["Timepoint"]
last_timepoint_ramicane = ramicane_only.groupby(["Mouse ID"]).max()["Timepoint"]
last_timepoint_infubinol = infubinol_only.groupby(["Mouse ID"]).max()["Timepoint"]
last_timepoint_ceftamin = ceftamin_only.groupby(["Mouse ID"]).max()["Timepoint"]

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
capomulin_merged = pd.merge(last_timepoint_capomulin, clean_mice, on=("Mouse ID","Timepoint"))
capomulin_merged.head()
```

```python
ramicane_merged = pd.merge(last_timepoint_ramicane, clean_mice, on=("Mouse ID","Timepoint"))
ramicane_merged.head()
```

```python
infubinol_merged = pd.merge(last_timepoint_infubinol, clean_mice, on=("Mouse ID","Timepoint"))
infubinol_merged.head()
```

```python
ceftamin_merged = pd.merge(last_timepoint_ceftamin, clean_mice, on=("Mouse ID","Timepoint"))
ceftamin_merged.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes

    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    
capomulin_quartiles = capomulin_merged.quantile([.25,.5,.75])["Tumor Volume (mm3)"]
capomulin_lowerq = capomulin_quartiles[0.25]
capomulin_upperq = capomulin_quartiles[0.75]
capomulin_iqr =  capomulin_upperq-capomulin_lowerq

capomulin_upper_bound = capomulin_upperq + (1.5*capomulin_iqr)
capomulin_lower_bound = capomulin_lowerq - (1.5*capomulin_iqr)

capomulin_max = capomulin_merged.max()["Tumor Volume (mm3)"]
capomulin_min = capomulin_merged.min()["Tumor Volume (mm3)"]

print("The IQR for Capomulin is " + str(capomulin_iqr) + ".")
print("Potential outliers for Capomulin are any values above " + str(capomulin_upper_bound)
      + " and any values below " + str(capomulin_lower_bound) + ".")
if capomulin_max > capomulin_upper_bound:
    print(str(capomulin_max) + " is an outlier for Capomulin.")
elif capomulin_min < capomulin_lower_bound:
    print(str(capomulin_min) + " is an outlier for Capomulin.")
else:
    print("There are no outliers for Capomulin.")
```
